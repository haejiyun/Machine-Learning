### Nettoyage des données disponibilité

```python
# Vérification des doublons

dispo.duplicated().sum()
```

```python
# Vérification des variables numériques

display(dispo.describe())
```

```python
# Modalités des variables qualitatifs - disponibilité alimentaire

display(dispo['Domaine'].value_counts()), display(dispo['Élément'].value_counts()), 
display(dispo['Produit'].value_counts()), display(dispo['Unité'].value_counts()), 
display(dispo['Symbole'].value_counts()), display(dispo['Description du Symbole'].value_counts())
```

```python
# Extraction des données utiles

dis = dispo[['Zone','Élément','Produit','Valeur', 'Unité']]
dis = dis.rename(columns = {'Valeur':'Disponibilité'})
dis.head()
```

```python
# Extraction d'unité

unit = dis[['Élément','Unité']].drop_duplicates()
unit
```

```python
# Pivot

dis = dis.pivot(index = ['Zone', 'Produit'], columns = 'Élément', values = 'Disponibilité').reset_index()
dis.head()
```

```python
# Vérification des modalités de variable produit

dis.Produit.unique()
```

```python
# Extraction de volailles

dis = dis[dis.Produit == "Viande de Volailles"].drop('Produit', axis = 1)
dis.head()
```

```python
# Vérification d'importance de chaque colonnes

dis.iloc[:,1:].sum(axis = 0)
```

```python
# Différence disponibilité intérieure - nourriture

dispo_nourriture = dis.loc[:,('Disponibilité intérieure','Nourriture')]
dispo_nourriture['différence'] = dispo_nourriture['Disponibilité intérieure'] - dispo_nourriture['Nourriture']
dispo_nourriture[dispo_nourriture['différence'] > 500]
```

```python
dis.columns
```

```python
dis = dis[['Zone', 'Disponibilité alimentaire (Kcal/personne/jour)', 
           'Disponibilité alimentaire en quantité (kg/personne/an)',
           'Disponibilité de matière grasse en quantité (g/personne/jour)',
           'Disponibilité de protéines en quantité (g/personne/jour)',
           'Disponibilité intérieure', 'Exportations - Quantité','Importations - Quantité', 'Production']]
dis.head()
```

<h2><span style="color:indianred"> 2. Combinaison des données </span></h2>

### Concaténation des données

```python
# Merge

data = dis.merge(pop[pop.Année == 2007][['Zone','Population']], on = 'Zone')
data.head()
```

### Nettoyage des données concaténées

```python
# Info

data.info()
```

```python
# Suppression des pays qui n'ont pas de valeur disponibilité intérieure
# Remplacement des valeurs manquantes par 0

data = data.drop(data[data['Disponibilité intérieure'].isna()].index)
data.iloc[:,5:] = data.iloc[:,5:].fillna(data.iloc[:,5:].mean(axis=0))
data.info()
```

```python
data.head()
```

### Export de données

```python
data.to_csv('data.csv')
```

<h2><span style="color:indianred"> 3. Analyse Exploratoire </span></h2>

### Disponibilité alimentaire

```python
# Analyse de la disponibilité

dispo_personne = data.iloc[:,2:5]
dispo_personne['Disponibilité alimentaire en quantité (g/personne/jour)'] = data['Disponibilité alimentaire en quantité (kg/personne/an)'] * 1000 /365
dispo_personne.head()
```

```python
# Visualisation de variabilité disponibilité


fig = plt.figure(figsize=(20, 10))
plt.title('\nVariabilité de disponibilité alimentaire liée à la volaille\n - g/personne/an -\n', 
          size = 25, color ='dimgray')
plt.xticks([])
plt.yticks([])
plt.box(False)

fig.add_subplot(111)
plt.boxplot(dispo_personne.iloc[:,1:], patch_artist=True, vert = False,
            labels = ['Disponibilité proteine', 'Disponibilité matière grasse', 'Disponibilité alimentaire'],
            boxprops=dict(facecolor='darkolivegreen', color ='white'), whiskerprops=dict(color='darkolivegreen'),
            capprops=dict(color='darkolivegreen'), flierprops=dict(markeredgecolor='darkolivegreen'),
            medianprops=dict(color='white'), widths = 0.5)
plt.grid('whitegrid', color ='gray', alpha = 0.3)
plt.box(False)
plt.tick_params(labelsize=15);
```

### Production et commerce

```python
# Commerce et production

commerce = data.iloc[:,6:-1]
commerce.head()
```

```python
# Visualisation de commerce et production

fig = plt.figure(figsize=(20, 10))
plt.title('\nVariabilité de production et de commerce extérieur \n - milliers de tonnes -\n',  
          size = 25, color = 'dimgray')
plt.xticks([])
plt.yticks([])
plt.box(False)

fig.add_subplot(111)
plt.boxplot(commerce, patch_artist=True, labels = commerce.columns, vert=False,
            boxprops=dict(facecolor='indianred', color ='indianred'), whiskerprops=dict(color='indianred'),
            capprops=dict(color='indianred'), flierprops=dict(markeredgecolor='indianred'),
            medianprops=dict(color="white"), widths = 0.5)
plt.grid('whitegrid', color ='gray', alpha = 0.3)
plt.xlim(0,1500)
plt.box(False)
plt.tick_params(labelsize=15);
```

### Population

```python
# Visualisation de variabilité de population

fig = plt.figure(figsize=(20, 5))
plt.grid('whitegrid', color ='gray', alpha = 0.3)
plt.box(False)
plt.title('\nVariabilité de la population \n- 1000 personnes -\n', fontsize = 20, color = 'dimgray')
plt.boxplot(data.iloc[:,-1], patch_artist=True, labels = ['Population'], vert = False,
            boxprops=dict(facecolor='forestgreen', color ='forestgreen'), whiskerprops=dict(color='forestgreen'),
            capprops=dict(color='forestgreen'), flierprops=dict(markeredgecolor='forestgreen'),
            medianprops=dict(color="white"), widths = 0.3)
plt.xlim(0,200000)
plt.tick_params(labelsize=15);
```
